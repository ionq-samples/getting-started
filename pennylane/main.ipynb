```python
import pennylane as qml

import os
from getpass import getpass

# Get your API key from https://cloud.ionq.com/settings/keys
api_key = os.getenv('IONQ_API_KEY') or getpass('Enter your IonQ API key: ')
dev = qml.device(
    'ionq.simulator',
    api_key=api_key,
    wires=2,
    shots=1000
)
```

```python
@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=0)
    qml.CNOT(wires=[0, 1])
    return qml.counts()
```

```python
fig, ax = qml.draw_mpl(circuit, style='sketch')()
```

```python
results = circuit()
print(results)
```
